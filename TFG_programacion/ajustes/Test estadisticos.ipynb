```python
# Los datos no parecen seguir una distribución normal, aplicamos el test de Wilcoxon
p_valor_viajeros = []
for i in range(len(labels)-1):
    for j in range(i+1, len(labels)):
        print("Comparando viajeros ", labels[i], " y " , labels[j])
        _,aux = sc.stats.wilcoxon(df_viajeros[labels[i]],df_viajeros[labels[j]])
        print("p-valor {:.5f}".format(aux))
        p_valor_viajeros.append(aux)
```
